```python
#Apply Function

def fun(value):
    if value>3000:
        return 'Yes'
    else:
        return 'No'

data['newColumn'] = data['ApplicantIncome'].apply(fun)
data.head()
```

```python
#By using the lambda operator
data['outputColumn'] = data['LoanAmount'].apply(lambda x:x/10)
data['apColumn'] = data['ApplicantIncome'].apply(lambda x:x/10)
data.head()
```

```python
val1={'id':[1,2,10,12],
   'val1':['a','b','c','d']}
df1=pd.DataFrame(val1)

val2={'id':[1,2,9,8],
      'val1':['p','q','r','s']}
df2=pd.DataFrame(val2)

print(df1)
print(df2)

#inner join returns df having same/common values & natural join is inner join

df=pd.merge(df1,df2, on='id', how ='inner')
print(df)
```

```python
#left
df=pd.merge(df1,df2, on='id', how ='left')
print(df)
```

```python
#right
df=pd.merge(df1,df2, on='id', how ='right')
print(df)
```

```python
#outer
#left
df=pd.merge(df1,df2, on='id', how ='outer')
print(df)
```

```python
#outer
#left
df=pd.merge(df1,df2, left_index=True,right_index=True)
print(df)
```
